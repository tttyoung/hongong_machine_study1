```python
# 층 추가하는 방법1
from tensorflow import keras
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
from sklearn.model_selection import train_test_split
train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
# 입력층
inputs = keras.layers.Input(shape=(784,))
# 은닉층. 100개의 뉴런을 가지고 활성화 함수로 시그모이드 함수를 사용함.
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784,))
# 출력층
dense2 = keras.layers.Dense(10, activation='softmax')
# 심층 신경망 만들기
model = keras.Sequential([inputs, dense1, dense2])
# 모델정보
model.summary()

# 층 추가하는 방법2
# 클래스 생성자 내부에서 입력층,은닉층, 출력층 객체 바로 만듦
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(100, activation='sigmoid', name='은닉층'),
    keras.layers.Dense(10, activation='softmax', name='출력층')
], name='패션 MNIST 모델')
model.summary()

# 층 추가하는 방법3
# 클래스 객체의 메서드 add 활
model = keras.Sequential()
model.add(keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)))
model.add(keras.layers.Dense(10, activation='softmax'))
model.summary()

# 훈련
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
# 렐루 함수

model = keras.Sequential()
model.add(keras.layers.Input(shape=(28,28)))
# Flatten 클래스는 2차원->1차원으로 펼치는 역할.
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
# Flatten이 층이라고는 하지만 학습하는 층이 아니므로 신경망 깊이가 3인건 아님.
# 하지만 입력값의 차원을 알 수 있다는 장점이 있음.
model.summary()
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
train_scaled = train_input / 255.0
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
model.evaluate(val_scaled, val_target)
```

```python
# 옵티마이저

model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
# 옵티마이저를 Adam으로 설정
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
model.evaluate(val_scaled, val_target)
```
